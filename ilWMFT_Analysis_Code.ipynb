## Abstract
**Objective**: The purpose of this study was to examine if motor performance of the ipsilesional (IL) hand varies directly with motor impairment and performance of the contralesional (CL) hand, and if this relationship differs based on the lesioned hemisphere.<br><br>
**Hypothesis**: <br>
***A. The Null Hypothesis***, wherein the relationship between contralesional impairment and ipsilesional performance is not modified by the side of stroke lesion (i.e., LHD or RHD). <br>***B. Alternative Hypothesis 1***, wherein IL deficits are related to CL impairment but only in LHD and RHD with severe impairment (e.g., UEFM < 34). <br>***C. Alternate Hypothesis 2***, wherein IL deficits are related to CL impairment but only in LHD and not in RHD.

```r
x = rep(0:66,2)
y1 = c(rev(seq(from = 30, to = 50, length.out=20)),rep(30,47))
y2 = rev(seq(from = 0.75, to = 50, length.out=67))
y3 = rev(rep(25,67))
d = data.frame(x,y1,y2,y3)
library(ggplot2)

gg1 = ggplot(d, aes(x = x,y=y2)) +
  geom_line(alpha=0.9,size=2,col="red") + 
  geom_line(aes(y=y2+10),alpha=0.9,size=2,col="blue") + 
  xlab("CL Impairment") + ylab("IL Performance") + ggtitle("A. Null Hypothesis") +
  #theme_bw() + theme(panel.grid = element_blank())+
        theme(axis.text.x=element_blank(),axis.text.y=element_blank(),
        axis.ticks.x=element_blank(),axis.ticks.y=element_blank())

gg2 = ggplot(d, aes(x = x,y=y1)) +
  geom_line(alpha=0.9,size=2,col="red") + 
  geom_line(aes(y=y2+5),alpha=0.9,size=2,col="blue") + 
  xlab("CL Impairment") + ylab("IL Performance") + ggtitle("C. Alternate Hypothesis 2") +
        annotate("rect",xmin=-Inf, xmax=20, ymin=-Inf, ymax=Inf, alpha=0.2,fill="black") +
        theme(axis.text.x=element_blank(),axis.text.y=element_blank(),
        axis.ticks.x=element_blank(),axis.ticks.y=element_blank())

gg3 = ggplot(d, aes(x = x,y=y3)) +
  geom_line(alpha=0.9,size=2,col="red") + 
  geom_line(aes(y=y2),alpha=0.9,size=2,col="blue") + 
  xlab("CL Impairment") + ylab("IL Performance") + ggtitle("B. Alternate Hypothesis 1") +
        theme(axis.text.x=element_blank(),axis.text.y=element_blank(),
        axis.ticks.x=element_blank(),axis.ticks.y=element_blank())

options(repr.plot.width=7.5, repr.plot.height=2.5)
source('~/Documents/My Documents/USC/Coursework/BKN 599 Stats in R/Data & Exercises/Data Packages/multiplot.R')
multiplot(gg1,gg3,gg2,cols=3)
```

**Methods**: Participants were 42 pre-morbidly right-handed chronic stroke survivors (left hemisphere damage = 21) with mild-to-moderate paresis (upper extremity Fugl-Meyer range: 19-59). Motor performance for both hands was quantified using the distal items of the Wolf Motor Function Test time-score (dWMFT) and were log- transformed to meet assumptions for statistical testing. The relationship between motor performance of IL and CL hands was assessed using the co-efficient of determination ($R^{2}$) and multivariate linear regression.<br><br>
**Results**: Compared to RHD, the relationship between contralesional arm impairment (Upper Extremity Fugl-Meyer, UEFM) and ipsilesional hand motor capacity was stronger ($R^{2}_{LHD} =$ 0.42; $R^{2}_{RHD} =$ < 0.01; z = 2.12; p = 0.03) and the slope was steeper (t = -2.03; p = 0.04) in LHD. Similarly, the relationship between contralesional dWMFT and ipsilesional hand motor capacity was stronger ($R^{2}_{LHD} =$ 0.65; $R^{2}_{RHD} =$ 0.09; z = 2.45; p = 0.01) and the slope was steeper (t = 2.03; p = 0.04) in LHD compared to RHD. Multiple regression analysis confirmed the presence of an interaction between contralesional UEFM and side of stroke ($_{3}=$ 0.66  0.30; p = 0.024) but only trended towards significance for the interaction between contralesional dWMFT and side of stroke ($_{3}=$-0.51  0.34; p = 0.05). Results were confirmed after removal of potential outliers. <br><br> 

**Conclusions**: The severity of contralesional deficits directly influences ipsilesional hand motor performance in those with LHD but not RHD. Compared with individuals with LHD, those with RHD preserve a motoric advantage for the ipsilesional dominant hand and also demonstrate a disproportionate disadvantage for the contralesional non-dominant hand. These finding demonstrate that the relationship between motor performance of the contralesional and ipsilesional arm depend on the hemisphere of stroke.

### Comparing Slopes between LHD and RHD
**Note:** Again, we don't expect to see a difference in slopes in this model, i.e., between UEFM and CoDistAvg... This we expect to be the same between LHD and RHD

```r
# m1, m2: the sample means
# s1, s2: the sample standard deviations
# n1, n2: the same sizes
t.test2 <- function(m1,m2,s1,s2,n1,n2){
    se = sqrt( s1^2 + s2^2)
    t <- (m1-m2)/se
    df = (n1 + n2) - 4
    dat <- c(m1-m2, df, se, t, 2*pt(-abs(t),df))    
    names(dat) <- c("Difference in slopes", "Df","Std Error", "t", "p-value")
    return(dat) 
}
```

```r
display_markdown("<br>**T-Test to Compare Slopes between LHD and RHD -- all samples**")
t.test2(aL$coefficients[2],aR$coefficients[2],
        summary(aL)$coefficients[2,2],summary(aR)$coefficients[2,2],
        21,21)

display_markdown("<br>**T-Test to Compare Slopes between LHD and RHD -- w/o Outliers**")
t.test2(bL$coefficients[2],bR$coefficients[2],
        summary(bL)$coefficients[2,2],summary(bR)$coefficients[2,2],
        19,19)
```

### Multivariate Linear Model
**Model 0 Full:** $\;\;$ *CoDistAvg* $=  _{0} +\; _{1}$ (*UEFM*) $ + \;_{2}$ (*SoL*)  $ + \;_{3}$ (*UEFM * SoL*) $ + \; _{4}$ (*Age*) $+ \; _{5}$(*Sex*) $+\; _{6}$(*Chronicity*)$ \;+ \;e $ <br><br>
**Model 0 Reduced:** $\;\;$ *CoDistAvg* $=  _{0} +\; _{1}$ (*UEFM*) $ \;+ \;_{2}$ (*SoL*)  $ + \;_{3}$ (*UEFM * SoL*) $ + \;_{4}$ (*Age*) $+ \; _{5}$(*Sex*) $\; +\; e $ <br><br>
**Hypotheses:**<br>
$\: \:$ $H_{0}: _{1} = 0\;$ <br>
$\: \:$ $ H_{1}: _{1}  0\; $ <br>

**Comparing the Full and Reduced Models:** <br> We begin with a full model in which we include age, sex, chronicity, uefm, and side of lesion (as well as our hypothesized interaction term), all as predictors. Then, we remove those predictors that don't satisfy the p = 0.2 lenient criteria. Below are logical outputs from the full model, for those coefficients which meet this liberal cut-off. That is, the predictors for which the value is TRUE are those that the p-value cut-off was met. These predictors were included in the final reduced model.

```r
# Full Model with all explanatory variables for backward selection
lm.fullMod1 = lm(zln.CoDistAvg ~ z.uefm*sol + z.age + sex + zln.chron, data = wmft2)
summary(lm.fullMod1)$coefficients[,4]<0.2

# Reduced model after taking out zln.chron which was not statistically significant
lm.redMod1 = lm(zln.CoDistAvg ~ z.uefm*sol + z.age + sex, data = wmft2)
lm.redMod1_remOut = lm(zln.CoDistAvg ~ z.uefm*sol + z.age + sex, data = wmft2)

huxreg(Full = lm.fullMod1,Reduced = lm.redMod1,Reduced_NoOuts = lm.redMod1_remOut) #just for fun! 
                                                      #Not for comparison between w and w/o outlier models
```

#### Estimates from the Final Reduced Model & Model Diagnostics

```r
# Display reduced model estimates, diagnostics and plot model fit
tidy(lm.redMod1,conf.int = TRUE); glance(lm.redMod1)
options(repr.plot.width=12, repr.plot.height=3); par(mfrow=c(1,4))
plot(lm.redMod1);
```

### <font color = 'hotpink5'> Interim Summary </font> 
<font color = 'mediumvioletred'> The final reduced form of Model 0 was statistically different from a null model (F = 16.24, p < 0.0001, adjusted $R^{2}=$ 0.65). **Based on estimates from this model, CL impairment (UEFM) was significantly associated with CL hand motor performance, i.e., dWMFT, ($_{1} =$ -0.83  0.11, p < 0.0001).** However, there was no significant effect of the side of lesion ($_2 =$ -0.37  0.19, p = 0.05), or the interaction between CL impairment and side of lesion ($_3  =$ 0.35  0.21, p = 0.09). Nor were the confounding factors (age, chronicity, and sex) found to be statistically significant.

### Multivariate Linear Model
**Model 1 Full:** $\;\;$ *IpDistAvg* $=  _{0} +\; _{1}$ (*UEFM*) $ + \;_{2}$ (*SoL*)  $ + \;_{3}$ (*UEFM * SoL*) $ + \; _{4}$ (*Age*) $+ \; _{5}$(*Sex*) $+\; _{6}$(*Chronicity*)$ \;+ \;e $ <br><br>
**Model 1 Reduced:** $\;\;$ *IpDistAvg* $=  _{0} +\; _{1}$ (*UEFM*) $ + \;_{2}$ (*SoL*)  $ + \;_{3}$ (*UEFM * SoL*) $ + \; _{4}$(*Sex*) $\; +\; e $ <br><br>
**Hypotheses:**<br>
$\: \:$ $H_{0}: _{1} = 0\;$ & $\;_{3} = 0.$<br>
$\: \:$ $ H_{1}: _{1}  0\;$ & $ \;_{3}  0. $ <br><br>
*Note*: $\;_{3}$ is the interaction effect that we seek to test. This is the main hypothesis of this paper.<br>

**Comparing the Full and Reduced Models:** <br> We begin with a full model in which we include age, sex, chronicity, uefm, and side of lesion (as well as our hypothesized interaction term), all as predictors. Then, we remove those predictors that don't satisfy the p = 0.2 lenient criteria. Below are logical outputs from the full model, for those coefficients which meet this liberal cut-off. That is, the predictors for which the value is TRUE are those that the p-value cut-off was met. These predictors were included in the final reduced model.

```r
# Full Model with all explanatory variables for backward selection
lm.fullMod2 = lm(zln.IpDistAvg ~ z.uefm*sol + sex + z.age + zln.chron, data = wmft2)
summary(lm.fullMod2)$coefficients[,4]<0.2

# Reduced model after taking out zln.chron which was not statistically significant
lm.redMod2 = lm(zln.IpDistAvg ~ z.uefm*sol + sex, data = wmft2)
lm.redMod2_remOut = lm(zln.IpDistAvg ~ z.uefm*sol + sex, data = wmft2[-Outs,])

huxreg(Full = lm.fullMod2,Reduced = lm.redMod2,Reduced_NoOuts = lm.redMod2_remOut)
```
